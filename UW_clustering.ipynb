```python
filepath="people_wiki.csv"
data=pd.read_csv(filepath)
data.head()
```

# creating word_vector using scikit learn countVectorizer

```python
# (creating word_vector using scikit learn countVectorizer)

from sklearn.feature_extraction.text import CountVectorizer

texts=data[data['name']=='Elton John']["text"]
#texts=["this is is my name ak ak ak ak. in the forward"]
cv = CountVectorizer()
cv_fit=cv.fit_transform(texts)

vocab=cv.get_feature_names()
count=cv_fit.toarray()[0]

word_vector=pd.DataFrame(vocab,columns=["word"])
word_vector["count"]=count
word_vector.sort("count",ascending=False)
```

# Creating tf-idf using TfidfVectorizer

```python
# Creating tf-idf

from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
import string


def tokenize(text):
    tokens = nltk.word_tokenize(text)
    return tokens

token_dict = {}
x=0
def create_corpus(text):
    global x
    x=x+1
    text=str(text)
    lowers=text.lower()
    no_punctuation = lowers.translate(None, string.punctuation)
    token_dict[x] = no_punctuation

data["text"].apply(lambda x: create_corpus(x))

#this can take some time
tfidf = TfidfVectorizer(tokenizer=tokenize, stop_words='english')
tfs = tfidf.fit_transform(token_dict.values())
print len(token_dict)
print tfs

#idf = vectorizer.idf_
#print dict(zip(vectorizer.get_feature_names(), idf))
```

```python
Elton_John=data[data['name']=='Elton John']
temp=str(Elton_John["text"])
response = tfidf.transform([temp])
print response
```

```python
feature_names = tfidf.get_feature_names()

rowlist=[]
for col in response.nonzero()[1]:
    dic={}
    dic["word"]=feature_names[col]
    dic["tfidf"]=response[0, col]
    rowlist.append(dic)
#    print feature_names[col], ' - ', response[0, col]
df = pd.DataFrame(rowlist, columns=["word","tfidf"])
df.sort("tfidf",ascending=False)
```
